# Caso práctico

Se desea obtener información específica de la población en el estado de Querétaro. Para ello se utilizarán las bases de datos proporcionados por el CONEVAL e INEGI, específicamente, las líneas de bienestar y los estudios ENOA, ETOA y ENIGH.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
l_bienestar = pd.read_html('https://portalsocial.guanajuato.gob.mx/content/canasta-alimentaria', header=0)[1]
```

```python
columns = l_bienestar.iloc[0].to_list()
columns[2] = columns[2] + ' (rural)'
columns[3] = columns[3] + ' (urbana)'
columns[4] = columns[4] + ' (rural)'
columns[5] = columns[5] + ' (urbana)'

l_bienestar.columns = columns

l_bienestar = l_bienestar.drop([0, 1])
l_bienestar = l_bienestar.reset_index(drop=True)

for col in l_bienestar.columns[2:]:
    l_bienestar[col] = l_bienestar[col].str.replace("$", "")
    l_bienestar[col] = l_bienestar[col].str.replace(",", "").astype(float)
```

```python
l_bienestar
```

```python
bien_min_17 = (l_bienestar[l_bienestar['Año'] == "2017"]["Bienestar mínimo (canasta alimentaria) (rural)"].mean() + l_bienestar[l_bienestar['Año'] == "2017"]["Bienestar mínimo (canasta alimentaria) (urbana)"].mean()) / 2
bien_min_17
```

```python
bien_min_18 = (l_bienestar[l_bienestar['Año'] == "2018"]["Bienestar mínimo (canasta alimentaria) (rural)"].mean() + l_bienestar[l_bienestar['Año'] == "2018"]["Bienestar mínimo (canasta alimentaria) (urbana)"].mean()) / 2
bien_min_18
```

```python
for i, j in l_bienestar['Mes'].iteritems():
    if j == 'Ene':
        l_bienestar.loc[i, 'Mes'] = "01"
    elif j == 'Feb':
        l_bienestar.loc[i, 'Mes'] = "02"
    elif j == 'Mar':
        l_bienestar.loc[i, 'Mes'] = "03"
    elif j == 'Abr':
        l_bienestar.loc[i, 'Mes'] = "04"
    elif j == 'May':
        l_bienestar.loc[i, 'Mes'] = "05"
    elif j == 'Jun':
        l_bienestar.loc[i, 'Mes'] = "06"
    elif j == 'Jul':
        l_bienestar.loc[i, 'Mes'] = "07"
    elif j == 'Ago':
        l_bienestar.loc[i, 'Mes'] = "08"
    elif j == 'Sep':
        l_bienestar.loc[i, 'Mes'] = "09"
    elif j == 'Oct':
        l_bienestar.loc[i, 'Mes'] = "10"
    elif j == 'Nov':
        l_bienestar.loc[i, 'Mes'] = "11"
    elif j == 'Dic':
        l_bienestar.loc[i, 'Mes'] = "12"
```

```python
l_bienestar['Año'] = l_bienestar['Año'].astype(str)
l_bienestar['Fecha'] = l_bienestar['Año'] + "-" + l_bienestar['Mes']
l_bienestar['Fecha'] = pd.to_datetime(l_bienestar['Fecha'])
l_bienestar = l_bienestar.drop(['Año', 'Mes'], axis=1)
l_bienestar.to_csv('bienestar.csv', index=False, encoding='Latin-1')
```

```python
sdem_17_1 = pd.read_csv('sdemt117.csv', engine='python')
sdem_17_2 = pd.read_csv('conjunto_de_datos_sdem_enoe_2017_2t.csv', engine='python')
sdem_17_3 = pd.read_csv('conjunto_de_datos_sdem_enoe_2017_3t.csv', engine='python')
sdem_17_4 = pd.read_csv('conjunto_de_datos_sdem_enoe_2017_4t.csv', engine='python')
```
